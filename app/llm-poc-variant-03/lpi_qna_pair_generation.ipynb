```python
# Function to extract course details from a row
def extract_course_details(row):
    return {
        "module_code": row['Module_Code'],
        "course_title": row['Course_Learning_Material'],
        "source": row['Source'],
        "level": row['Course_Level'],
        "cost_type": row['Type_Free_Paid'],
        "duration": row['Duration'],
        "difficulty": row['Difficulty_Level'],
        "skills": row['Keywords_Tags_Skills_Interests_Categories'],
        "link": row['Links']
    }
```

```python
# Function to create a context string for QA generation
def create_context(details):
    return f"Course Title: {details['course_title']}; Module Code: {details['module_code']}; Level: {details['level']}; Type: {details['cost_type']}; Duration: {details['duration']}"
```

```python
# Function to define question templates with types and metadata
def generate_question_templates(details):
    return [
        # Factual Recall Questions
        {
            "instruction": f"What is the title of the course with Module Code '{details['module_code']}'?",
            "input": create_context(details),
            "output": details["course_title"],
            "metadata": {"type": "Factual Recall", "difficulty": details["difficulty"], "cognitive_level": "Recall", "source_reference": details["course_title"]}
        },
        {
            "instruction": f"Is the course '{details['course_title']}' free or paid?",
            "input": f"Module Code: {details['module_code']}; Cost Status: {details['cost_type']}",
            "output": details["cost_type"],
            "metadata": {"type": "Factual Recall", "difficulty": details["difficulty"], "cognitive_level": "Recall", "source_reference": details["course_title"]}
        },

        # Comprehension Questions
        {
            "instruction": f"Explain the main purpose of the course '{details['course_title']}' provided by {details['source']}.",
            "input": f"Course Source: {details['source']}; Skills Covered: {details['skills']}",
            "output": f"The course '{details['course_title']}' by {details['source']} focuses on {details['skills']}.",
            "metadata": {"type": "Comprehension", "difficulty": details["difficulty"], "cognitive_level": "Comprehension", "source_reference": details["course_title"]}
        },
        {
            "instruction": f"What are the primary skills or topics covered in the '{details['course_title']}' course?",
            "input": f"Keywords: {details['skills']}",
            "output": details["skills"],
            "metadata": {"type": "Comprehension", "difficulty": details["difficulty"], "cognitive_level": "Comprehension", "source_reference": details["skills"]}
        },

        # Application Questions
        {
            "instruction": f"How much time will it take to complete the '{details['course_title']}' course?",
            "input": f"Duration: {details['duration']}",
            "output": f"The estimated time to complete the course is {details['duration']}.",
            "metadata": {"type": "Application", "difficulty": details["difficulty"], "cognitive_level": "Application", "source_reference": details["duration"]}
        },
        {
            "instruction": f"Provide an example of a project where skills from '{details['course_title']}' would be useful.",
            "input": f"Skills: {details['skills']}",
            "output": f"Skills from '{details['course_title']}' could be applied in a project involving {details['skills']}, such as developing machine learning models.",
            "metadata": {"type": "Application", "difficulty": details["difficulty"], "cognitive_level": "Application", "source_reference": details["course_title"]}
        },

        # Analysis Questions
        {
            "instruction": f"Compare the '{details['course_title']}' course with other {details['level']} courses offered by {details['source']}.",
            "input": f"Source: {details['source']}; Level: {details['level']}",
            "output": f"The course '{details['course_title']}' is rated {details['difficulty']} and covers topics such as {details['skills']}, making it comparable to other {details['level']} courses by {details['source']}.",
            "metadata": {"type": "Analysis", "difficulty": details["difficulty"], "cognitive_level": "Analysis", "source_reference": details["course_title"]}
        },
        {
            "instruction": f"Where can you access the '{details['course_title']}' course online?",
            "input": f"Course Link: {details['link']}",
            "output": details["link"],
            "metadata": {"type": "Analysis", "difficulty": details["difficulty"], "cognitive_level": "Analysis", "source_reference": details["link"]}
        },

        # Synthesis/Evaluation Questions
        {
            "instruction": f"Evaluate the usefulness of '{details['course_title']}' for learners interested in {details['skills']}.",
            "input": f"Level: {details['level']}; Skills: {details['skills']}; Difficulty: {details['difficulty']}",
            "output": f"The course '{details['course_title']}' by {details['source']} is beneficial for learners interested in {details['skills']}, providing material at a {details['difficulty']} level.",
            "metadata": {"type": "Evaluation", "difficulty": details["difficulty"], "cognitive_level": "Evaluation", "source_reference": details["course_title"]}
        },
        {
            "instruction": f"How could the course '{details['course_title']}' be improved for {details['level']} learners?",
            "input": f"Course Level: {details['level']}; Duration: {details['duration']}",
            "output": f"The course '{details['course_title']}' could be improved with hands-on projects to enhance practical understanding for {details['level']} learners.",
            "metadata": {"type": "Synthesis", "difficulty": details["difficulty"], "cognitive_level": "Synthesis", "source_reference": details["course_title"]}
        }
    ]
```

```python
# Main function to generate QA pairs
def generate_qa_pairs(row):
    details = extract_course_details(row)
    questions = generate_question_templates(details)
    qa_pairs = [{"instruction": q["instruction"], "input": q["input"], "output": q["output"], "metadata": q["metadata"]} for q in questions]
    return qa_pairs
```

```python
# Generate Q&A pairs for each row in the dataset and combine them into a single list
all_qa_pairs = []
for _, row in df.iterrows():
    all_qa_pairs.extend(generate_qa_pairs(row))

# Convert the list of Q&A pairs into a DataFrame for further analysis or export
qa_pairs_df = pd.DataFrame(all_qa_pairs)

# Optional: Save the generated Q&A pairs to a CSV file
qa_pairs_df.to_csv("qa_pairs.csv", index=False)
# print("Q&A pairs generated and saved to 'Generated_QA_Pairs_Enhanced.csv'")
```
